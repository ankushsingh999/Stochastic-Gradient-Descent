```python
def fmseMiniBe(X, y, w, b):
    p = np.dot(X.T,w)
    d = (p-y)
    FMSEB = np.mean(d)
    return FMSEB

def FMSE(X, y,w,b):
 
    a = np.square(np.dot(X.T, w) + b - y)
    FMSEW = (np.mean(a)/(2))
    return FMSEW

def fmseMiniWe(X, y, w, b, a):
    #with respect to weights
    bias = (np.dot(X.T,w))+b-y
    error = (np.dot(X,bias))/X.shape[1]
    L = (a*w)/X.shape[1]
    FMSEWE= error + L
    return FMSEWE

def SGD (x, y,w, b, e, lr, a , mini):
    for i in range(e):
        batches = int((len(x.transpose())/mini))
        start = 0
        end = mini
        for j in range(batches):
            
            xminib = x[:,start:end]
            yminib = y[start:end]
            #print("sixe xd:",xminib.shape)
            
            
            #fmseMiniB = np.mean(np.dot(xminib.transpose(),w)+b - yminib)
            #with respect to w
            fmseMiniB = fmseMiniBe(xminib,yminib,w,b)
            fmseMiniW = fmseMiniWe(xminib,yminib,w,b,a)
            
            wnew = w - (np.dot(lr, fmseMiniW))
            bnew = b - (np.dot(lr, fmseMiniB))
            #print("hfihjihfih")
            start = end
            end = end+mini
        fmse = np.mean(np.square(np.dot(x.transpose(),wnew)+bnew-y))
    return [wnew, bnew, fmse]          
            



def train_age_regressor ():
    # Load data
    X_tr = np.reshape(np.load("age_regression_Xtr.npy"), (-1, 48*48))
    ytr = np.load("age_regression_ytr.npy")
    X_te = np.reshape(np.load("age_regression_Xte.npy"), (-1, 48*48))
    yte = np.load("age_regression_yte.npy")

    #divivding the data set
    X = X_tr.transpose()
    print(X.shape)
    indx = np.random.permutation(X.shape[1])
    XTrain = X[:,indx[:int(X.shape[1]*0.8)]]
    XValid = X[:,indx[int(X.shape[1]*0.8):]]
    yTrain = ytr[indx[:int(X.shape[1]*0.8)]]
    yTrain = np.atleast_2d(yTrain).T
    yValid = ytr[indx[int(X.shape[1]*0.8):]]
    yValid = np.atleast_2d(yValid).T

    print("train:",XTrain.shape)
    print(yTrain.shape)
    print("valid:",XValid.shape)
    print(yValid.shape)
    
    #random initial weights
    w = np.random.randint(-10,10,int(X.shape[0]))
    w = np.atleast_2d(w).T
    #considering initial bias as 0
    b = 0
    print(w.shape)
    
    e = [80,90,100,250]
    lr = [0.0001,0.0002,0.0003,0.0005]
    a = [0.1,3,2,5]
    mini = [30,40,50,100]
    
    #intialising the Cost function to have an extremely large value
    FMSEinitial = np.random.uniform(100000000 , 1000000000)
    for ne in e:
        for nlr in lr:
            for na in a:
                for nmini in mini:
                    w,b,fmse = SGD(XTrain,yTrain,w,b,ne,nlr,na,nmini)
                    fmseValid = FMSE(XValid,yValid,w,b)
                    if fmseValid<FMSEinitial:
                        print("epochs ", ne," Learning rate ",nlr,"alpha ", na," mini batch",nmini)
                        FMSEinitial = fmseValid
                        print("Minimum FMSE",FMSEinitial)
                        enew = e
                        lrnew = lr
                        anew = a
                        mininew = mini
    
                  
    wTrain, bTrain, fmseTrain = SGD(XTrain,YTrain,enew, lrnew, anew, mininew)
    
    #for test set 
    
    fmseTest = np.mean(np.square(np.dot(X_te.transpose(),yte, wTrain,b)))
    print("FMSE for Test:  ",fmseTest)
    
    
    
    
    #w,b = linear_regression(X_tr, ytr)
   
    

    
    
```
